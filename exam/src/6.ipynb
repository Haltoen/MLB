```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data
d = np.loadtxt('visualization-data-MLB2023.csv', delimiter=",")

# 3d plot
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(d[:, 0], d[:, 1], d[:, 2], c='blue', label='Points')

# Set axis labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.tight_layout()
ax.legend()
plt.title('Three-dimensional Scatterplot of Dataset')
plt.show()
```

```python
from sklearn.decomposition import PCA


# Perform PCA on normalized data
pca = PCA(n_components=2)
pca_2d_res = pca.fit_transform(d)

# Plot resulting 2d points
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.scatter(pca_2d_res[:,0], pca_2d_res[:, 1], c='blue', label='Transformed data')
plt.title('PCA on Unnormalized Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
```

```python
from sklearn.manifold import TSNE

perplexities = [10, 50, 200]
fig, axes = plt.subplots(3, 1, figsize=(6, 12))
fig.suptitle('t-SNE with Different Perplexity Values')

# run t-SNE and produce plots
for i, perplexity in enumerate(perplexities):
    tsne = TSNE(n_components=2, perplexity=perplexity, n_iter=1000, random_state=42)
    d_tsne = tsne.fit_transform(d)

    # Plot t-SNE results
    ax = axes[i ]
    ax.scatter(d_tsne[:, 0], d_tsne[:, 1], c='blue', label='Transformed data')
    ax.set_title(f't-SNE with Perplexity={perplexity}')
    ax.set_xlabel('Component 1')
    ax.set_ylabel('Component 2')
    ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
